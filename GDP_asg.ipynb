```python
import glob, os
```

```python
# Path where all the files are stored
path = r'C:\Users\sakusuma\Desktop\pgdds\GDP_Assignment\data1b' # use your path
# Reading all the paths
all_files = glob.glob(path + "/*.csv")

# Empty list to add all the dataframes
li = []
```

```python
# Looping to read each file with the path given in the filename variable.
for filename in all_files:
    # Reading each file as a pandas dataframe
    df_1b = pd.read_csv(filename, index_col=None, header=0,encoding='ANSI')
    # Selecting only required columns
    df_1b = df_1b[['Item','2014-15']]
    # Transpose the data
    df_1b = df_1b.T
    # Taking the header row
    new_header = df_1b.iloc[0] #grab the first row for the header
    df_1b = df_1b[1:] #take the data less the header row
    # Assign the new header
    df_1b.columns = new_header 
    # Add the section name from the filename
    df_1b.loc[:,'State'] = filename.split('\\')[-1].split('-')[1]
    # Append the final DF to the list
    li.append(df_1b)
```

```python
# Creating a single dataframe by joining all the df_1bs in the list li
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
#removing the Union Territories from the dataframe.
filter1 = frame['State'] == "Chandigarh"
filter2 = frame['State'] == "Delhi"
filter3 = frame['State'] == "Puducherry"
filter4 = frame['State'] == "Andaman & Nicobar Islands"
filter9 = frame['State'] == "Dadra & Nagar Haveli"
filter10 = frame['State'] == "Daman & Diu"

frame = frame[~(filter1 | filter2 | filter3 | filter4 | filter9 | filter10)]
```

```python
#Soritng the dataframe on the Per capita GDP
frame.sort_values(by='Per Capita GSDP (Rs.)', ascending = False, inplace = True)
```

```python
#caluculating the Primary sector Percentage in the total GSDP
frame['%Primary'] = frame['Primary']*100/frame['Gross State Domestic Product']
```

```python
#caluculating the Secondary sector Percentage in the total GSDP
frame['%Secondary'] = frame['Secondary']*100/frame['Gross State Domestic Product']
```

```python
#caluculating the Primary Tertiary Percentage in the total GSDP
frame['%Tertiary'] = frame['Tertiary']*100/frame['Gross State Domestic Product']
```

```python
#caluculating the total percentages of all the sectors
frame['%TotSec'] = frame['%Primary'] + frame['%Secondary'] + frame['%Tertiary']
```

```python
#caluculating the Taxes collected Percentage in the total GSDP
frame['%Taxes'] = frame['Taxes on Products']*100/ frame['Gross State Domestic Product']
```

```python
#caluculating the Subsidies Percentage in the total GSDP
frame['%Subsidies'] = frame['Subsidies on products']*100/frame['Gross State Domestic Product']
```

```python
#Displaying all the Series together so that we can see what is the math bewtween all of them.
frame[['%Primary','%Secondary', '%Tertiary', '%TotSec', '%Taxes', '%Subsidies']]
```

### Why is (Primary + Secondary + Tertiary) not equal to total GDP?
Ans: If we observe in the above result. %TotalSec(i.e Primary + Secondary + Tertiary) + %Taxes -%Subsidies = 100%
As the there are taxes being collected and also part of the income to the nation, so just the sectors contribution wont be the total GDP
